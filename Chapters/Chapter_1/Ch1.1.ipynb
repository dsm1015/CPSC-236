## Functions
Function
: piece of prewritten code that performs an operation

`print` function
: displays output on the screen

Argument 
: data given to a function

Statements in a program execute in the order that they appear from top to bottom

## Comments
Comments 
: Notes of explanation within a program
```python
# This comment gets ignored by the interpreter
print("This will print")
```

End-line comment
: appears at the end of a line of code
```python
z = x + y # This line adds x and y and stores it in z
```

## Variables
Variable
: name that represents a value stored in the computer memory

Assignment statement
: used to create a variable and make it reference data
```python
age = 22
```

### Data Types
Data Types
: categorize value in memory

Variables can be assigned to a different type:
```python
age = 22
age = "twenty-two"
```

**Variables naming rules:**
- Variable name cannot be a Python key word
- Variable name cannot contain spaces
- First character must be a letter or an underscore
- After first character may use letters, digits, or
underscores
- Variable names are case sensitive
- *Variable name *should* reflect its use

## Reading Input

`input` function returns the user data as a string
```python
age = input("Enter age:")
print(age) # "22"
```

Convert to int data type:
```python
age = int(input("Enter age:"))
print(age) # 22
```

Python operator precedence:
1. Operations enclosed in parentheses
2. Exponentiation (**)
3. Multiplication (*), division (/ and //), and remainder (%)

*Higher precedence performed first*

*Same precedence operators execute from left to right*


# In Class 1
Write a program that obtains hours and minutes from seconds.

````python
# in class 1

seconds = int(input("seconds:"))
hours = seconds//3600
minutes = (seconds - 3600*hours)//60
rSeconds = seconds - hours*3600 - minutes*60
print("Hours: ", hours)
print("Minutes:", minutes)
print("Seconds:", rSeconds)
````

# In Class 2
Write a program that converts a Fahrenheit degree to Celsius using the formula:

`celsius = (5/9)(fahrenheit - 32)`

````python
# in class 2
far = float(input("Far: "))
cel = round((5/9)*(far-32),2)
print(cel)
````

 # Magic Number and Named Constants

Avoid unexplained numeric values in your code.

Dont do this:
